```python
from statsmodels.tsa.arima.model import ARIMA

# Example ARIMA(1,1,1) model
model_arima = ARIMA(combined_df['Total Environmental Intensity (Revenue)'], order=(1,1,1))
result_arima = model_arima.fit()

# Print model summary
print(result_arima.summary())
```

```python
# Make predictions
y_pred_arima = result_arima.predict(start=len(X_train), end=len(X_train)+len(X_test)-1, typ='levels')

from sklearn.metrics import mean_squared_error, mean_squared_log_error , mean_absolute_error
# Evaluate the ARIMA model
mse_arima = mean_squared_error(y_test, y_pred_arima)
rmse_arima = np.sqrt(mse_arima)
mae_arima = mean_absolute_error(y_test, y_pred_arima)

print(f"ARIMA MSE: {mse_arima:.4f}")
print(f"ARIMA RMSE: {rmse_arima:.4f}")
print(f"ARIMA Mean Absolute Error: {mae_arima:.4f}")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
combined_df['Total Environmental Intensity (Revenue)'] = scaler.fit_transform(
    combined_df['Total Environmental Intensity (Revenue)'].values.reshape(-1, 1)
)
```

```python
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

# Choose a sequence length (number of time steps to look back)
sequence_length = 10

# Create sequences for training
X_seq, y_seq = create_sequences(combined_df['Total Environmental Intensity (Revenue)'].values, sequence_length)
```

```python
# Split the data into training and testing sets
split_ratio = 0.8
split_index = int(split_ratio * len(X_seq))

X_train, X_test = X_seq[:split_index], X_seq[split_index:]
y_train, y_test = y_seq[:split_index], y_seq[split_index:]
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(units=50, input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Reshape the input data to be 3D [samples, timesteps, features]
X_train_reshaped = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

# Train the model
model.fit(X_train_reshaped, y_train, epochs=50, batch_size=32)
```

```python
# Reshape the input data for prediction
X_test_reshaped = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Make predictions on the test set
y_pred_lstm = model.predict(X_test_reshaped)

# Invert the scaling to get predictions in the original scale
y_pred_lstm = scaler.inverse_transform(y_pred_lstm)
y_test_inverse = scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluate the LSTM model
mse_lstm = mean_squared_error(y_test_inverse, y_pred_lstm)
rmse_lstm = np.sqrt(mse_lstm)
mae_lstm = mean_absolute_error(y_test_inverse, y_pred_lstm)

print(f"LSTM MSE: {mse_lstm:.4f}")
print(f"LSTM RMSE: {rmse_lstm:.4f}")
print(f"LSTM Mean Absolute Error: {mae_lstm:.4f}")
```
